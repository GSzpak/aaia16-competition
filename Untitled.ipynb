```python
from __future__ import division

from pymongo import MongoClient

from src.import_data import DB_NAME

with MongoClient() as client:
    db = client[DB_NAME]
    group_by_clauses = [
        [{
            "$group": {
                "_id": {"label": "$label"},
                "count": {"$sum": 1},
            }
        }],
        [{
            "$group": {
                "_id": {"working_id": "$main_working_id", "label": "$label"},
                "count": {"$sum": 1}
            }
        }],
        [{
            "$group": {
                "_id": {"file": "$file", "label": "$label"},
                "count": {"$sum": 1},
            }
        }],
        [{
            "$group": {
                "_id": {"file": "$file", "working_id": "$main_working_id", "label": "$label"},
                "count": {"$sum": 1},
            }
        }],
    ]
    print "LABELS SUMMARY:"
    collection = db.training_data
    for group_by_clause in group_by_clauses:
        result = list(collection.aggregate(group_by_clause))
        group_by = ", ".join(result[0]["_id"].keys())
        print group_by
        temp_result = {}
        for elem in result:
            label = elem["_id"].pop("label")
            groupby_id = ", ".join(map(str, elem["_id"].values()))
            if groupby_id in temp_result:
                temp_result[groupby_id][label] = elem['count']
            else:
                temp_result[groupby_id] = {label: elem['count']}
        for groupby_id, labels in sorted(temp_result.items()):
            normal_count = labels.get("normal", 0)
            warn_count = labels.get("warning", 0)
            all_ = normal_count + warn_count
            ratio = warn_count / normal_count if normal_count > 0 else 0
            print "\t{0}\tall: {1}, warning: {2}, normal: {3}, ratio: {4:.4f}".format(groupby_id, all_, warn_count, normal_count, ratio)
        print 
```
